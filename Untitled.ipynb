```python
import requests
import os
import pandas as pd
```

```python
def download_data(url, save_path):
    response = requests.get(url)
    
    if response.status_code == 200:
        with open(save_path, 'wb') as file:
            file.write(response.content)
        print("Data downloaded successfully!")
    else:
        print(f"Failed to download data. Status code: {response.status_code}")
```

```python
# Replace 'YOUR_URL_HERE' with the actual URL of the WHO dataset you want to download
url = 'https://data.world/data-society/health-status-indicators'
```

```python
# Replace 'your_save_path_here.csv' with the desired local file path to save the downloaded data
save_path = 'data/health.csv'

# Download the data
download_data(url, save_path)
```
